```python
import numpy as np

#  Activation Function 
def tanh(x):
    return np.tanh(x)

# Initialize Weights
def initialize_weights():
    np.random.seed(42)  # to make results reproducible
    
    W1 = np.random.uniform(-0.5, 0.5, (2, 2))
    W2 = np.random.uniform(-0.5, 0.5, (2, 1))
    
    return W1, W2

# Forward Propagation 
def forward_propagation(X, W1, W2, b1, b2):
    
    # Hidden layer
    Z1 = np.dot(X, W1) + b1
    A1 = tanh(Z1)
    
    # Output layer
    Z2 = np.dot(A1, W2) + b2
    output = tanh(Z2)
    
    return A1, output

# Main 
if __name__ == "__main__":
    
    # Input
    X = np.array([[0.6, 0.1]])
    
    # Bias
    b1 = 0.5
    b2 = 0.7
    
    # Initialize weights
    W1, W2 = initialize_weights()
    
    # Forward pass
    hidden_output, final_output = forward_propagation(X, W1, W2, b1, b2)
    
    print("Hidden Layer Output:\n", hidden_output)
    print("Final Network Output:\n", final_output)
```
